```python
import load_data
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
DATASET_DIR = './bigdata_datasets'
datasets = load_data.get_datasets(DATASET_DIR)
print(datasets[0].datalist)
print(datasets[0].label)
print(len(datasets))
```

```python
labels = [i.label for i in datasets]
print(labels)
```

```python
[ (d.datalist.shape, d.label) for d in datasets ]
```

```python
def plot_spec_label(datasets, index):
    PLOT_INDEX = index
    NUM_COL = 8
    PLOT_LABEL = datasets[PLOT_INDEX].label
    plt.figure(PLOT_LABEL, figsize=(20,3))
    for i in range(NUM_COL):
        fig_name = str(PLOT_LABEL)+"col_"+str(i)
        plt.subplot(1, NUM_COL, i+1)
        for d in datasets:
            if d.label == PLOT_LABEL:
                try:
                    plt.plot(range(len(d.datalist[i])), d.datalist[i])
                except IndexError:
                    break;
        plt.title(fig_name)
```

```python
def plot_all():
    pre_label = ""
    for i in range(len(datasets)):
        if datasets[i].label != pre_label:
            plot_spec_label(datasets, i)
            pre_label = datasets[i].label
```

```python
plot_all()
```
