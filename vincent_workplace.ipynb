```python
def plot_spec_label(datasets, index, prefix="", ylim=None):
    PLOT_INDEX = index
    NUM_COL = 8
    PLOT_LABEL = datasets[PLOT_INDEX].label
    plt.figure(prefix+PLOT_LABEL, figsize=(20,3))
    for i in range(NUM_COL):
        fig_name = str(PLOT_LABEL)+"col_"+str(i)
        plt.subplot(1, NUM_COL, i+1)
        for d in datasets:
            if d.label == PLOT_LABEL:
                try:
                    plt.plot(range(len(d.datalist[i])), d.datalist[i])
                except IndexError:
                    break;
        if ylim:
            plt.ylim(ylim)
        plt.title(fig_name)
```

```python
def plot_all(datasets, prefix="", ylim=None):
    pre_label = ""
    for i in range(len(datasets)):
        if datasets[i].label != pre_label:
            plot_spec_label(datasets, i, prefix=prefix, ylim=ylim)
            pre_label = datasets[i].label
```

```python
plot_all(datasets)
```

```python
def differentiate(arr):
    arr_len = len(arr)
    return arr[1:arr_len] - arr[0:arr_len-1]

def non_zero_fix(datasets):
    MIN_LIM_TEMP = 20
    for d in datasets:
        for col in d.datalist:
            for i in range(len(col)):
                if col[i] < MIN_LIM_TEMP:
                    #Find front nonzero
                    #print(col)
                    temp_i = i
                    while temp_i > 0 and col[temp_i] < MIN_LIM_TEMP:
                        temp_i -= 1
                    col[i] = col[temp_i]
    return datasets

def interpolation_fix(datasets):
    for d in datasets:
        for col in d.datalist:
            offset = differentiate(col)
            for i in range(len(offset)):
                if abs(offset[i]) > 10:
                    try:
                        col[i+1] = (col[i] + col[i+2])/2
                    except IndexError:
                        col[i+1] = col[i]
    return datasets

def show_noise(datasets):
    new_d = []
    for d in datasets:
        new_cols = []
        newDataset = load_data.dataset()
        for col in d.datalist:
            offset = differentiate(col)
            new_cols.append(offset)
        newDataset.datalist = np.array(new_cols)
        newDataset.label = d.label
        
        new_d.append(newDataset)    
    return np.array(new_d)
```

```python
#datatsets = non_zero_fix(datasets)
fixed_d = non_zero_fix(datasets)
fixed_d = interpolation_fix(fixed_d)
plot_all(fixed_d)
noise = show_noise(fixed_d)
plot_all(noise, prefix="noise", ylim=(-15, 15))

#plot_all(fixed_d, ylim=(-15, 15))
```
